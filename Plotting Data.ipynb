In this notebook, we try to gain a deeper understanding about the distribution of data of the apps that we scraped in our previous step (Refer documentation if you've not run the ***play_store_scraper.py*** before running this notebook.


Here, we load the scraped data and perform the following operations

- Create a table for the top 10 categories (based on the number of installs) for each country

Based on the above 2 operations, we can choose the apps appropriately.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read the json file from the output folder

```python
try:
    data = pd.read_json('output/installs_per_category_for_100.json').T
except ValueError as e:
    print('Make sure you have run play_store_scraper.py before running this notebook. For more information refer Readme.')
```

#### We create a dataframe in which each category has the number of downloads (in 1000s) for the countries - India (in) and US (us)

```python
data
```

#### We plot a bargraph of app category v/s the number of downloads (in 10^10) for the countries (IN and US) as shown below

```python
ax = data.plot(figsize=(15,10),kind='bar', title='App Category v/s No. of downloads (10^10)', legend='best', grid=True)
ax.set_xlabel('Category')
ax.set_ylabel('Number of downloads (in 10^10)')
```

#### Let's find the mean of the data

```python
data.mean()
```

#### We find the top 10 app categories ranked by the number of downloads for Indian Playstore

```python
data.nlargest(11, ['in'])
```

#### We find the top 10 app categories ranked by the number of downloads for US Playstore

```python
data.nlargest(11, ['us'])
```

#### The top 10 apps based on the different markets using pandas.nlargest - [nlargest documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html)

***The order of the countries is really <span style='color:red'>important</span>. Refer the documentation above***
