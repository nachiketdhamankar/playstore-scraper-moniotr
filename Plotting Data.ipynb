#### Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Read the json file from the output folder

```python
data = pd.read_json('output/installs_per_category_for_100_rounded.json').T
```

### We create a dataframe in which each category has the number of downloads (in 1000) and the category

```python
data
```

#### We plot a bargraph of app category v/s the number of downloads (in 10^10) as shown below

```python
ax = data.plot(figsize=(15,10),kind='bar', title='App Category v/s No. of downloads (10^10)', legend='best', grid=True)
ax.set_xlabel('Category')
ax.set_ylabel('Number of downloads (in 10^10)')
```

```python
data.mean()
```

#### We find the top 10 app categories ranked by the number of downloads for Indian Playstore

```python
data.nlargest(11, ['in'])
```

#### We find the top 10 app categories ranked by the number of downloads for US Playstore

```python
data.nlargest(11, ['us'])
```

#### The top 10 apps based on Indian market first and then US market - [nlargest documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html)

```python
data.nlargest(10, ['in', 'us'])
```
